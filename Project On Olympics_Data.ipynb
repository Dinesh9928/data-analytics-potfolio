#### 2. Which country has won the most men's gold medals in Athletics over the years?. Sort the results alphabetically by the player's names.

```python
Gold_medals = df[(df['Medal']=='Gold') & (df['Gender']=='Men')&(df['Sport']=='Athletics')]
Gold_medals
```

```python
mens = Gold_medals.sort_values(by = 'Athlete')
```

```python
mens['NOC'].value_counts()[:15]
```

```python
len(mens['NOC'].value_counts())
```

#### 5. Which five countries have won the most medals in recent years (from 2000 to 2008)?m

```python
rec_medals = df[(df['Edition'] >= 2000)]
rec_medals
```

```python
rec_medals['NOC'].value_counts().head()
```

```python
recent_medals = rec_medals.sort_values(by='Edition', ascending=False)
recent_medals
```

```python
recent_medals['NOC'].value_counts().head(n=5)
```

#### 6. Display the male gold medal winners for the hockey event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
df['Event'].value_counts().head()
```

```python
male_gold_hundred = df[(df.Gender == 'Men') & (df.Medal == 'Gold') & (df.Event == 'hockey')]
male_gold_hundred
```

```python
male_gold_hundred.sort_values(by='Athlete', ascending = False)
```

```python
male_gold_hundred.sort_values(by = 'Edition', ascending=False )[['City','Edition','Athlete','NOC']]
```

#### 7. How many women's have won the gold medal's till 2008 in Olympic game

```python
women_gold = df[(df.Medal == 'Gold') & (df.Gender == 'Women')]
women_gold
```

```python
women_gold.sort_values(by ='Athlete')
```

```python
# Many women's have won more than one gold medals in the series of Olympic games, so let's get the total number of women's 
# with the help of value_counts() method
```

```python
women_gold['Athlete'].drop_duplicates()
```

```python
women_gold.sort_values(by ='Athlete', ascending=True )['Athlete'].value_counts()
```
